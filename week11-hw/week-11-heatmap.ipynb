```python
## Step 4: heatmap of bins through out all sample days
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
abund_file = '/Users/cmdb/qbb2020-answers/week11-hw/abundance_table.tab'
abundance = pd.read_csv(abund_file, sep = '\t')
abundance
```

```python
Bin_file = '/Users/cmdb/qbb2020-answers/week11-hw/week13_data/bins/all-bin-top-hit.txt' 
bin_labels = pd.read_csv(Bin_file, sep = '\t', names = ['Genomic bins', 'Species name'])
bin_labels
```

```python
ab_genomes = bin_labels.merge(abundance, on ='Genomic bins')
ab_genomes
```

```python
values = ab_genomes.columns.values[2:]
heatmap_data = pd.pivot_table(ab_genomes, values=values, 
                              index=['Genomic bins', 'Species name'])
```

```python
heatmap_data = np.log2(heatmap_data)
```

```python
sns.clustermap(heatmap_data, figsize=(14,10), col_cluster=False,cmap="coolwarm")
```
