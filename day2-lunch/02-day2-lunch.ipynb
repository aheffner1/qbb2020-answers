# Identifier mapping

```python
# Set up Function
def idmapping(ctab, mapping, field = 'not ignore'): 
    
    # initialize map_dict as dictionary and counter variable
    map_dict = dict()
    counter = -1
    
    # loop through each line in mapping file, strip and split line save in map_list as list
    # add map_list column 0 as key to map_dict and column 1 as values
    for line in mapping:
        map_list = list(line.strip().split())
        map_dict[map_list[0]] = map_list[1]
    
    # loop through each line in ctab file 
    for line in ctab:
        
        ctab_list = list(line.strip().split()) #strip and split line save in ctab_list as list
        gene_id = ctab_list[8]     # save ctab column 8 as gene_id
        
        if counter == -1: # remove header line
            counter += 1
            continue
        elif counter == 100: # continue for only 100 lines
            break
            
        # if gene_id is in the map dictionary then replace gene_id in ctab_list with 
        # uniprot or value of map_dict for that gene_id and print out statements
        if gene_id in map_dict:
            ctab_list[8] = map_dict[gene_id]
            print('Ctab with uniprot id:   ', ctab_list)
            print('Map of flybase id and uniprot id:   ', gene_id,map_dict[gene_id])
            print('Counter: ', counter,'\n')
            
        # else = gene_id is not in map dictionary
        else: 
            # look at field argument
            if field == 'ignore':   # if field is 'ignore' skip the rest
                continue
                
            # if field is not ignore and counter is greater then 1 (to remove header)
            # then print there is no map and counter
            elif field == 'not ignore' and counter >=0 :
                ctab_list[8] = 'Not Found'
                print('Ctab:   ', ctab_list)
                print('There is no map for:  ', gene_id)
                print('Counter', counter, '\n')   
        # add to count
        counter += 1
```

```python
# load files
mapping = open('/Users/cmdb/qbb2020-answers/day2-lunch/fly_genetoprot.txt', 'r')
ctab = open('/Users/cmdb/data/results/stringtie/SRR072893/t_data.ctab')
```

```python
# call function
idmapping(ctab, mapping)
```
